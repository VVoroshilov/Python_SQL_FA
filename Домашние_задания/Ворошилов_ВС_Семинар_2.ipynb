## В папке с файлом этого юпитер-ноутбука должна быть папка Data, в которой находится json файл

```python
import json
import os
path = os.getcwd() + os.path.sep + 'Data/data-7361-2022-01-19.json'
with open(path, encoding = "CP1251") as json_file:
    data = json.load(json_file)
```

# AdmArea - административный округ. 
# Структура: [{ObjectAddress : [{AdmArea }]}]

```python
# Проверка, что структура определена верно: 
data[0]["ObjectAddress"][0]["AdmArea"]
```

# Заведём два словаря
# Один для подсчёта количества библиотек в каждом административном округе 
# Второй для группирвоки. При группировке будем добавлять только название библиотеки. 
# Если хотим сохранить туда информацию, то заменим одну строчку. Она будет указана в комментарии в коде. 

```python
amount_AmdArea = dict()
amdArea_libs = dict()
```

```python
# Подсчитываем количество библиотек в округе 
for i in range(len(data)):
    AdmArea_name = data[i]["ObjectAddress"][0]["AdmArea"]
    if AdmArea_name not in amount_AmdArea.keys():
        amount_AmdArea[AdmArea_name] = 1
        amdArea_libs[AdmArea_name] = [data[i]["CommonName"]]
        # Если хотим сохранять всю информацию: 
        #amdArea_libs[AdmArea_name] = [data[i]]
    else:
        amount_AmdArea[AdmArea_name] += 1
        amdArea_libs[AdmArea_name].append(data[i]["CommonName"])
        # Если хотим сохранять всю информацию: 
        # amdArea_libs[AdmArea_name].append([data[i]])
```
