## 1. Предположим, что стипендия всем студентам увеличена на 20%. Напишите запрос к таблице STUDENT, выполняющий вывод номера студента, его фамилии и величины увеличенной стипендии. Выходные данные упорядочите: а) по значению последнего столбца (величине стипендии); б) в алфавитном порядке фамилий студентов.

```python
sql = """
SELECT student_id, surname, name, stipend*1.2 as 'Increased stipend'
FROM student
ORDER BY 4, 2;
"""
df = pd.read_sql(sql, conn)
df
```

## 2. Напишите запрос, который по таблице EXAM_MARKS позволяет найти а) максимальные и б) минимальные оценки каждого студента и выводит их вместе с идентификатором студента.

```python
sql = """
SELECT student_id, MAX(mark) AS "Best mark", MIN(mark) as "Worst mark"
FROM exam_marks
GROUP BY STUDENT_ID;
"""
df = pd.read_sql(sql, conn)
df
```

## 3. Напишите запрос, выполняющий вывод списка предметов обучения в порядке а) убывания семестров и б) возрастания количества отводимых на предмет часов. Поле семестра в выходных данных должно быть первым, за ним должны следовать имя предмета обучения и идентификатор предмета.

```python
sql = """
SELECT semester, hour, subj_name, subj_id
FROM subject
ORDER BY semester DESC;
"""
df = pd.read_sql(sql, conn)
df
```

```python
sql = """
SELECT semester, hour, subj_name, subj_id
FROM subject
ORDER BY hour;
"""
df = pd.read_sql(sql, conn)
df
```

## 4. Напишите запрос, который для каждой даты сдачи экзаменов выполняет вывод суммы баллов всех студентов и представляет результаты в порядке убывания этих сумм.

```python
sql = """
SELECT DATE(exam_date) AS 'Date of exam', sum(mark) AS 'Summary'
FROM exam_marks
GROUP BY exam_date
ORDER BY 2 DESC;
"""
df = pd.read_sql(sql, conn)
df
```

## 5. Напишите запрос, который для каждой даты сдачи экзаменов выполняет вывод а) среднего, б) минимального, в) максимального баллов всех студентов и представляет результаты в порядке убывания этих значений.

```python
order_param = 2
sql = f"""
SELECT DATE(exam_date) AS 'Date of exam', ROUND(AVG(mark), 3) AS 'Average', MIN(mark) AS 'Minimum', MAX(mark) AS 'Maximum'
FROM exam_marks
GROUP BY exam_date
ORDER BY {order_param} DESC;
"""
df = pd.read_sql(sql, conn)
df
```

```python
from sqlalchemy import MetaData, Table, create_engine, select, and_, or_, not_
from sqlalchemy.sql import func
```

```python
engine = create_engine('sqlite:////Users/vladislavvorosilov/Desktop/Python_SQL/Python_SQL_PI_2021_22_s2/Data/Students.db')
metadata_obj = MetaData() # Объект-контейнер, содержит информацию о схеме данных
metadata_obj.reflect(bind=engine) # Отображение БД
# Вызов метода reflect() приводит к включению в объект metadata_obj словаря всех таблиц БД.
```

```python
# 1 

# func. - вызов функций(например max, min, avg)
# table.c.column_name - синтаксис столбца

# Подзапрос
subquery = select(func.max(student.c.stipend)).scalar_subquery()
# Запрос
s = select(student.c.surname, student.c.name, student.c.stipend)
s = s.where(student.c.stipend == subquery)
s = s.order_by(student.c.surname)

result = conn.execute(s).fetchall()
print(s)
result[0:5]
```

```python
# 2
# С помощью подзапроса напишите запрос, выводящий список студентов, получающих стипендию, превышающую среднее значение стипендии.
subquery = select(func.avg(student.c.stipend)).scalar_subquery()
# Запрос
s = select(student.c.surname, student.c.name, student.c.stipend)
s = s.where(student.c.stipend > subquery)
s = s.order_by(student.c.surname)

result = conn.execute(s).fetchall()
print(f"Average = {conn.execute(func.avg(student.c.stipend)).fetchall()[0][0]}")
result[0:10]
```

```python
#3 
# С помощью подзапроса напишите запрос,
# выводящий список студентов, обучающихся в Воронеже, с последующей сортировкой по идентификаторам университетов и курсам.

subquery = select(university.c.univ_id).where(university.c.city == "Воронеж").scalar_subquery()
# Запрос
s = select(student.c.surname, student.c.name, student.c.univ_id, student.c.kurs)
s = s.where(student.c.univ_id.in_(subquery))
s = s.order_by(student.c.univ_id, student.c.kurs)

result = conn.execute(s).fetchall()
result[0:10]
```
