## 17. Для каждого университета напишите запрос, выводящий количество обучающихся в нем студентов, с последующей сортировкой списка по этому количеству.

```python
# Благодаря джоину с таблицей университетов, удалось узнать, что в таблице университетов имеется МФТИ, 
# однако, в таблице студентов нет ни одного студента, учащегося в МФТИ. Так мы добыли информацию, что в МФТИ учится 0 человек. 
sql = """
SELECT UNIVERSITY.UNIV_ID, UNIVERSITY.UNIV_NAME, COUNT(DISTINCT STUDENT_ID) AS AMOUNT
FROM UNIVERSITY 
LEFT JOIN STUDENT
ON (STUDENT.UNIV_ID = UNIVERSITY.UNIV_ID)
GROUP BY STUDENT.UNIV_ID
ORDER BY AMOUNT DESC
"""
df = pd.read_sql(sql, conn)
df
```

```python
# Вот пример запроса, который, скорее всего, сделают большинство студентов. Видно, что отсуствует информация о МФТИ (univ_id = 12)
sql = """
SELECT STUDENT.UNIV_ID, COUNT(STUDENT_ID) AS AMOUNT
FROM STUDENT 
GROUP BY STUDENT.UNIV_ID
ORDER BY AMOUNT DESC
"""
df = pd.read_sql(sql, conn)
df
```

## 18. Для каждого университета напишите запрос, выводящий количество работающих в нем преподавателей, с последующей сортировкой списка по этому количеству.

```python
# Ситуация аналогична с прошлым заданием. Благодаря джоину с таблицей университетов, мы получаем информацию об отсуствии 
# преподавателей в МФТИ. 
sql = """
SELECT UNIVERSITY.UNIV_ID, UNIVERSITY.UNIV_NAME, COUNT(DISTINCT LECTURER_ID) AS AMOUNT
FROM UNIVERSITY 
LEFT JOIN LECTURER
ON (LECTURER.UNIV_ID = UNIVERSITY.UNIV_ID)
GROUP BY LECTURER.UNIV_ID
ORDER BY AMOUNT DESC
"""
df = pd.read_sql(sql, conn)
df
```

## 25. Напишите запрос, выполняющий вывод списка студентов, имеющих только отличные оценки. Результат отстортируйте по идентификатору студента.

```python
# Критерия отличной оценки в задании не задан. Я задам её параметром в переменную rate_min и rate_max
# В своём решении я буду считать, что отличная оценка - только "5", 
# поэтому свой промежуток я задам отрезком от 5 до 5, то есть только точка 5 
# Такой подход особенно хорошо подойдёт для 100-балльных систем оценивания
rate_low = 5
rate_max = 5
sql = f"""
SELECT STUDENT_ID, MIN(MARK) AS 'Наименьшая оценка'
FROM EXAM_MARKS 
GROUP BY STUDENT_ID 
HAVING MIN(MARK) BETWEEN {rate_low} and {rate_max}
ORDER BY STUDENT_ID 
"""
df = pd.read_sql(sql, conn)
df
```
